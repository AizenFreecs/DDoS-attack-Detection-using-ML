```python
old_value = 'malicious'
new_value = 1
df['label'] = df['label'].replace(old_value, new_value)
old_value = 'benign'
new_value = 0
df['label'] = df['label'].replace(old_value, new_value)
```

```python
train, test = train_test_split(df, test_size=0.3)
```

```python
df.info()
```

```python
# function to save the cleaned dataset
def save_cleaned_dataset(dataframe: pd.DataFrame,dataset: str, tag: str = ""):
    # create a new directory to save the cleaned dataset
    os.makedirs('./Dataset/dataset_final', exist_ok=True)

    if not(tag == ""):
        tag = "_" + tag

    dataframe.to_csv(f'Dataset/dataset_final/{dataset}{tag}.csv', index=False)
```

```python
save_cleaned_dataset(dataframe=train, dataset='LUFlow_Train')
save_cleaned_dataset(dataframe=test, dataset='LUFlow_Test')
```
